````python
# Q1: Please try and write code for the following Matplotlib related prioblems
# i: X is variable, whose value varies from 0 – 20. Please plot 3 lines for equations y=X, y=X2 and y= X3.
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 20, 100)
plt.plot(x, x, label='y=x')
plt.plot(x, x**2, label='y=x^2')
plt.plot(x, x**3, label='y=x^3')
plt.legend()
plt.show()

#ii: Create a pie chart and bar diagram with the following sample data
      #Sample data:
      #Programming languages: Java, Python, PHP, JavaScript, C#, C++
      #Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
y = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.bar(x, y)
plt.show()
plt.pie(y, labels=x)
plt.show()

# randomly generated mean age and std deviation across 3 classes in one plot
x = np.random.randint(18, 30, 50)
y = np.random.randint(18, 30, 50)
z = np.random.randint(18, 30, 50)

class_data = [x, y, z]
means = [np.mean(x), np.mean(y), np.mean(z)]
stds = [np.std(x), np.std(y), np.std(z)]

positions = [1, 2, 3]
plt.figure(figsize=(12, 6))
plt.ylim(10, 40)
plt.boxplot(class_data, positions=positions, tick_labels=["Class A", "Class B", "Class C"])

plt.errorbar(positions, means, yerr=stds, fmt="o", capsize=6, label="Mean ± 1 standard deviation")

plt.xlabel("Class")
plt.ylabel("Age")
plt.title("Age Distribution Across Three Classes")
plt.legend()
plt.grid(axis="y", axis="x")
plt.show()
````

````python
#Q2: Please perform the following activities on the penguin dataset.
#i: Import the dataset into pandas dataframe.
import pandas as pd
df = pd.read_csv('penguinsDT.csv')
pd.set_option('display.max_rows', 10)
print(df)

#ii:Add title for each column.
df.columns = ['species', 'island', 'bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'sex']
pd.set_option('display.max_rows', 10)
print(df)

#iii: Calculate the number of missing values in each column.
missing_values = df.isna().sum()
print(missing_values)

#iv: Draw a bar diagram showing the number of penguin species.
species_counts = df['species'].value_counts()
plt.bar(species_counts.index, species_counts.values)

#v: Create a scatter plot to present the relationship between Flipper Length and Body Mass.
plt.scatter(df['flipper_length_mm'], df['body_mass_g'], s=3)
plt.xlabel('Flipper Length')
plt.ylabel('Body Mass')
plt.xlim(150, 250)
plt.ylim(2500, 6500)
plt.show()

#vi: Create a bar diagram to present the average bill length of each species.
plt.bar(df['species'].unique(), df.groupby('species')['bill_length_mm'].mean())
plt.xlabel('Species')
plt.ylabel('Average Bill Length')
plt.show()

````
