````python
# Q1: Please try and write code for the following Matplotlib related prioblems
# i: X is variable, whose value varies from 0 – 20. Please plot 3 lines for equations y=X, y=X2 and y= X3.
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 20, 100)
plt.plot(x, x, label='y=x')
plt.plot(x, x**2, label='y=x^2')
plt.plot(x, x**3, label='y=x^3')
plt.legend()
plt.show()

#ii: Create a pie chart and bar diagram with the following sample data
      #Sample data:
      #Programming languages: Java, Python, PHP, JavaScript, C#, C++
      #Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
y = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.bar(x, y)
plt.show()
plt.pie(y, labels=x)
plt.show()

# randomly generated mean age and std deviation across 3 classes in one plot
x = np.random.randint(18, 30, 50)
y = np.random.randint(18, 30, 50)
z = np.random.randint(18, 30, 50)

class_data = [x, y, z]
means = [np.mean(x), np.mean(y), np.mean(z)]
stds = [np.std(x), np.std(y), np.std(z)]

positions = [1, 2, 3]
plt.figure(figsize=(12, 6))
plt.ylim(10, 40)
plt.boxplot(class_data, positions=positions, tick_labels=["Class A", "Class B", "Class C"])

plt.errorbar(positions, means, yerr=stds, fmt="o", capsize=6, label="Mean ± 1 standard deviation")

plt.xlabel("Class")
plt.ylabel("Age")
plt.title("Age Distribution Across Three Classes")
plt.legend()
plt.grid(axis="y", axis="x")
plt.show()
````

````python
#Q 1:Please perform the following activities on the CreditCard Fraud dataset.
#I. Import the dataset into pandas dataframe.
#II. Transform Non-numerical data types into numerical.
#III. Find the correlation between the features of the dataset
#IV. Split the dataset into training and testing dataset.
#V. Apply Logistic Regression Machine learning algorithm and generate the performance.
#VI. Apply decision tree and generate the performance.

import pandas as pd
from sklearn.metrics import confusion_matrix
dfp = pd.read_csv('payment_fraud.csv')
#print(dfp)

df = pd.get_dummies(dfp, columns = ['paymentMethod'])
#print(df)

import seaborn as sns
import matplotlib.pyplot as plt
dfcorr = df.corr()
#print(dfcorr)
sns.heatmap(dfcorr, annot = True)
plt.show()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('label', axis=1), df['label'], test_size=0.999, random_state=17)

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, classification_report
clf = LogisticRegression()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
print('Accuracy score:', accuracy_score(y_test, y_pred))
print('\n')
print('Confusion Matrix', '\n', confusion_matrix(y_test, y_pred))
print('\n')
print('Classification Report', '\n', classification_report(y_test, y_pred))

````

````python
#Q2: Please perform the following activities on the penguin dataset.
#I. Import the dataset into pandas dataframe.
#II. Transform Non-numerical data types into numerical using Encoding.
#III. Find the correlation between the features of the dataset
#IV. Split the dataset into training and testing dataset.
#V. Apply Logistic Regression Machine learning algorithm.
#VI. Print classification report

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

df = pd.read_csv("penguinswithextraCell.xls")
df = df.drop(columns=["Unnamed: 0"])
df = df.dropna()
print(df.isna().sum())

X = df.drop(columns=["species"])
y = df["species"]

X = pd.get_dummies(X, columns=["island", "sex"], drop_first=True, dtype=int)
print(X)

dfcorr = X.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(dfcorr, annot=True, fmt=".2f")
plt.title("Correlation Between Penguin Features")
plt.show()

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=17,
    stratify=y
)

clf = LogisticRegression(max_iter=2000)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

print("Accuracy score:", accuracy_score(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))


````
